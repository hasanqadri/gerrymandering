```python
import json
import csv
import pandas as pd
import numpy as np

df = pd.read_csv('house_results.csv', usecols=['State', 'District','Dem Total', 'Dem %', 'Dem Candidate', 'Gop Total', 'Gop %', 'Gop Candidate', 'Winner']);
states = {}

for i, row in enumerate(df.values):
    if (row[0] not in states):
        states[row[0]] = {}
    
for i, row in enumerate(df.values):
    states[row[0]][row[1]] = {
                                'Dem Total': row[2],
                                'Dem %': row[3],
                                'Dem Candidate': row[4],
                                'Gop Total': row[5],
                                'Gop %': row[6],
                                'Gop Candidate': row[7],
                                'Winner': row[8]
                                }
       
```

```python
print(states)
```

```python

pd.DataFrame(states).to_json("file_paths.json")

#np.savetxt("houseFormat1.csv", d, delimiter=",")
```
